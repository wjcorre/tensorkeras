# **Uso de datasets de para predicciones en una red neuronal**
Es util para implementar algoritmos de aprendizaje automatico, con el objetivo de entrenar redes neuronales

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
import tensorflow.keras as ke
import matplotlib.pyplot as plt
````

**Cargamos los datos con el mudulo pandas**

````python
train= pd.read_csv('/content/drive/MyDrive/Colab Notebooks/datasetstitanic/train.csv')
train.head()
````

**El modelo se trata de predecir si una persona sobrevivio o no a la inundacion del barco del titanic**

En ese caso la variable de partida sera Survived

````python
train_parametro_base= train['Survived']
````

A continuación especificamos las variables necesarias para el proceso de entrenamiento de a red

````python
datos_columnas_relevantes=['Pclass','Sex','Age','Cabin','Fare','Embarked']
````

````python
train_relevantes=train[datos_columnas_relevantes]
````

````python
train_relevantes.head()
````

# **Aplicacion del One hot encoding**

Generar el "one hot encoding" para la columna "Pclass"

````python
train_relevantes= pd.concat([train_relevantes, pd.get_dummies(train_relevantes['Pclass'], prefix='Pclass')],axis=1);
train_relevantes.drop(['Pclass'],axis=1, inplace=True)
````

Generar el "one hot encoding" para la columna "Sex"

````python
train_relevantes= pd.concat([train_relevantes, pd.get_dummies(train_relevantes['Sex'], prefix='Sex')],axis=1);
train_relevantes.drop(['Sex'],axis=1, inplace=True)
````

Generar el "one hot encoding" para la columna "Embarked"

````python
train_relevantes= pd.concat([train_relevantes, pd.get_dummies(train_relevantes['Embarked'], prefix='Embarked')],axis=1);
train_relevantes.drop(['Embarked'],axis=1, inplace=True)
````

````python
train_relevantes.head()
````

````python
train_relevantes['Cabin'].unique()
````

# **Eliminamos valores nulos**

````python
train_relevantes['Cabin']= train_relevantes['Cabin'].fillna('C')
````

Agregamos el valor inicial de cada vector en la columna

````python
train_relevantes['Cabin_1']= train_relevantes['Cabin'].astype('str').str[0]
````

````python
train_relevantes
````

Elaboramos el metodo de one hot encoding nuevamente con la intencion de limpiar la nueva columna 'Cabin_1'

````python
train_relevantes= pd.concat([train_relevantes, pd.get_dummies(train_relevantes['Cabin_1'], prefix='Cabin_1')],axis=1);
train_relevantes.drop(['Cabin_1'],axis=1, inplace=True)
train_relevantes.drop(['Cabin'],axis=1, inplace=True)
train_relevantes.drop(['Age'],axis=1, inplace=True)
````

Comprobamos el resultado de la matriz

````python
train_relevantes
````

Eliminamos las columnas que a partir de la limpieza quedan totalmente en nulidad

````python
train_relevantes.drop(['Cabin_1_T'],axis=1, inplace=True)
````

````python
train_relevantes
````

Comprobamos que el numero de filas no hayan sido alteradas

````python
train_parametro_base
````

Como se observa todo es correcto, ahora esta listo para la elaboración de la red neuronal

#**Uso del modulo Keras**

Inicializamos nuestra red neuronal

````python
modelo= ke.Sequential([
                       ke.layers.Input((None,16,)),
                       ke.layers.Dense(64,activation='relu'),
                       ke.layers.Dropout(0.2),
                       ke.layers.Dense(128,activation='relu'),
                       ke.layers.Dense(1,activation='sigmoid')

])
````

Compilamos el modelo

````python
modelo.compile(optimizer='adam',
               loss=ke.losses.binary_crossentropy,
               metrics=['accuracy']
               )
````

Comprobamos si el modelo se encuentra en aprendizaje, y al mismo tiempo entrenado

````python
historico=modelo.fit(train_relevantes,train_parametro_base,validation_split=0.2,epochs=100)
````

Iniciamos la evaluacion

````python
test_loss, tess_acc=modelo.evaluate(train_relevantes,train_parametro_base,verbose=1)
````

````python
print(historico.history.keys())
````
